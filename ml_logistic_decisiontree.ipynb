```python
import pandas as pd
```

```python
df=pd.read_csv("Heart.csv")
df
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
le
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc
```

```python
df["ChestPain"]=le.fit_transform(df["ChestPain"])
df
```

```python
df["Thal"]=le.fit_transform(df["Thal"])
df
```

```python
df["AHD"]=le.fit_transform(df["AHD"])
df
```

```python
x=df.drop("AHD",axis=1)
x
```

```python
y=df["AHD"]
y
```

```python
x.dropna()
y.dropna()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3)
```

```python
xtrain_sc=sc.fit_transform(xtrain)
```

```python
xtest_sc=sc.fit_transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
lo=LogisticRegression(random_state=0)
```

```python
lo.fit(xtrain_sc,ytrain)
```

```python
p=lo.predict(xtest_sc)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ytest,p)*100
acc
```

```python
from sklearn.tree  import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=3,min_samples_split=5,min_samples_leaf=3)
dt.fit(xtrain_sc,ytrain)
```

```python
w=dt.predict(xtest_sc)
w
```

```python
ww=accuracy_score(ytest,w)*100
ww
```

```python
from sklearn import tree
tree.plot_tree(dt)
```
