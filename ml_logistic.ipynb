```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("scr-dataset.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

```

```python
df["ChestPain"]=le.fit_transform(df["ChestPain"])
df["Thal"]=le.fit_transform(df["Thal"])
df["AHD"]=le.fit_transform(df["AHD"])
df=df.dropna()
```

```python
x=df.drop("AHD",axis=1)
df
```

```python
x.head()
```

```python
y=df["AHD"]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3,random_state=45)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
xtrain_scaled=scaler.fit_transform(xtrain)
xtest_scaled=scaler.fit_transform(xtest)
xtrain_scaled
xtest_scaled
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0)
```

```python
model=lr.fit(xtrain_scaled,ytrain)
ypred=model.predict(xtest_scaled)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ypred,ytest)*100
acc
```
